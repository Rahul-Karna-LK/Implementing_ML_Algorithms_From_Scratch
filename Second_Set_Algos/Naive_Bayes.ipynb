```python
def true_positive(ground_truth, prediction):
    tp = 0
    for gt, pred in zip(ground_truth, prediction):
        if gt == 1 and pred == 1:
            tp +=1
    return tp

def true_negative(ground_truth, prediction):
    tn = 0
    for gt, pred in zip(ground_truth, prediction):
        if gt == 0 and pred == 0:
            tn +=1
    return tn

def false_positive(ground_truth, prediction):
    fp = 0
    for gt, pred in zip(ground_truth, prediction):
        if gt == 0 and pred == 1:
            fp +=1
    return fp

def false_negative(ground_truth, prediction):
    fn = 0
    for gt, pred in zip(ground_truth, prediction):
        if gt == 1 and pred == 0:
            fn +=1
    return fn

def metrics(y, y_pred):
    y = [1 if i==' >50K' else 0 for i in y]
    y_pred = [1 if i==' >50K' else 0 for i in y_pred]
    tp = true_positive(y, y_pred)
    tn = true_negative(y, y_pred)
    fp = false_positive(y, y_pred)
    fn = false_negative(y, y_pred)
#     print(tp, tn, fp, fn)
    accuracy = (tp + tn)/ (tp + tn + fp + fn)
    print("Accuracy: %f" % (accuracy))
    precision = tp/ (tp + fp) 
    print("Precision: %f" % (precision))
    recall = tp/ (tp + fn)
    print("Recall: %f" % (recall))
    f1 = 2*precision*recall/(precision+recall)
    print("F1 score: %f" % (f1))
```

## Smoothing Methods

### Laplace Smoothing
This technique involves adding a small constant value to the count of each feature/class combination in the training set. This ensures that even rare events get a non-zero probability estimate. The smoothing parameter is usually set to 1, hence the name "add-one" smoothing.

```python
class NaiveBayesClassifier:
  priorProb = []
  classes = []
  class_counts = []
  attribute_values = []
  attribute_class_probs = []
  alpha = 1 # for Laplace smoothing
  def __init__(self):
    pass

  def calcClass(self, daf): 
    self.classes, self.class_counts = np.unique(daf['income'], return_counts=True)

  #calculate prior probabilities
  def calc_prior_probabilities(self, y):
    unique, counts = np.unique(y, return_counts=True)
    counts = counts/len(y)
    data = np.array([unique, counts])
    dataset = pd.DataFrame({'Class': data[0, :], 'PriorProbability': data[1, :]})
    dataset.PriorProbability = dataset.PriorProbability.astype(float)
    self.priorProb = dataset

  #finding all probabilities in each attribute
  def calc_conditional_prob(self, daf):
    for i in range(0, 14):
      unique, counts = np.unique(daf.iloc[:, i], return_counts=True) 
      self.attribute_values.append(unique)

    for i in range(0, 14):
      att_p = []
      for j in range(0, len(self.attribute_values[i])):
        probs = []
        for k in range(0, len(self.classes)):
          df1 = daf[daf.iloc[:, i] == self.attribute_values[i][j]]
          df2 = df1[df1['income'] == self.classes[k]]
          probs.append((df2.shape[0] + self.alpha) / (self.class_counts[k] + self.alpha * len(self.attribute_values[i])))
        att_p.append(probs)
      self.attribute_class_probs.append(att_p)

  def predict(self, ser):
    prob = []
    for i in range(0, len(self.classes)):
      p = self.priorProb.at[i, 'PriorProbability']
      for a in range(0, 14):
        for b in range(0, len(self.attribute_values[a])):
          if ser[a] == self.attribute_values[a][b]:
            p *= self.attribute_class_probs[a][b][i]
      prob.append(p)
    
    maxi = 0
    maxval = 0
    for i in range(0, len(self.classes)):
      if prob[i] > maxval:
        maxi = i
        maxval = prob[i]
    return self.classes[maxi]

  def calc_df_classes(self, daf):
    pred = []
    for i in range(0, len(daf)):
      pred.append(self.predict(daf.iloc[i]))
    return pred

  def calc_accuracy(self, pred, actual):
    count = 0
    for i in range(0, len(pred)):
      if pred[i] == actual[i]:
        count = count + 1
    return count / len(pred)
```

### Lidstone smoothing
This is a generalization of Laplace smoothing, where the smoothing parameter is set to a value between 0 and 1. The formula for computing the smoothed probability of a feature/class combination is:

P(f | c) = (count(f, c) + alpha) / (count(c) + alpha * V)

where count(f, c) is the count of feature f in class c, count(c) is the total count of class c, V is the total number of possible feature values, and alpha is the smoothing parameter.

```python
class NaiveBayesClassifier:
    priorProb = []
    classes = []
    class_counts = []
    attribute_values = []
    attribute_class_probs = []
    alpha = 1  # default value for Lidstone smoothing
    
    def __init__(self, alpha=1):
        self.alpha = alpha
    
    def calcClass(self, daf):
        self.classes, self.class_counts = np.unique(daf['income'], return_counts=True)
        
    # calculate prior probabilities
    def calc_prior_probabilities(self, y):
        unique, counts = np.unique(y, return_counts=True)
        counts = counts/len(y)
        data = np.array([unique, counts])
        dataset = pd.DataFrame({'Class': data[0, :], 'PriorProbability': data[1, :]})
        dataset.PriorProbability = dataset.PriorProbability.astype(float)
        self.priorProb = dataset
    
    # find all probabilities in each attribute
    def calc_conditional_prob(self, daf):
        for i in range(0, 14):
            unique, counts = np.unique(daf.iloc[:, i], return_counts=True) 
            self.attribute_values.append(unique)
            
        for i in range(0, 14):
            att_p = []
            for j in range(0, len(self.attribute_values[i])):
                probs = []
                for k in range(0, len(self.classes)):
                    df1 = daf[daf.iloc[:, i] == self.attribute_values[i][j]]
                    df2 = df1[df1['income'] == self.classes[k]]
                    probs.append((df2.shape[0] + self.alpha) / (self.class_counts[k] + len(self.attribute_values[i]) * self.alpha))
                att_p.append(probs)
            self.attribute_class_probs.append(att_p)
            
    def predict(self, ser):
        prob = []
        for i in range(0, len(self.classes)):
            p = self.priorProb.at[i, 'PriorProbability']
            for a in range(0, 14):
                found = False
                for b in range(0, len(self.attribute_values[a])):
                    if ser[a] == self.attribute_values[a][b]:
                        found = True
                        p = p * self.attribute_class_probs[a][b][i]
                        break
                if not found:
                    p = p * (self.alpha / (self.class_counts[i] + len(self.attribute_values[a]) * self.alpha))
            prob.append(p)

        maxi = 0
        maxval = 0
        for i in range(0, len(self.classes)):
            if prob[i] > maxval:
                maxi = i
                maxval = prob[i]
        return self.classes[maxi]

    def calc_df_classes(self, daf):
        pred = []
        for i in range(0, len(daf)):
            pred.append(self.predict(daf.iloc[i]))
        return pred

    # accuracy
    def calc_accuracy(self, pred, actual):
        count = 0
        for i in range(0, len(pred)):
            if pred[i] == actual[i]:
                count = count + 1
        return count / len(pred)
```

### Dirichlet smoothing
In probability theory, the Dirichlet distribution is a family of continuous probability distributions with parameters that can be used to model a distribution over a finite set of categorical outcomes. By incorporating a Dirichlet prior distribution over the parameters of a multinomial distribution, Dirichlet smoothing estimates the probability of unseen events by adding a small positive constant alpha to each count, effectively "smoothing out" the probabilities and reducing the impact of zero counts.

Dirichlet smoothing helps to avoid overfitting of the model and improves the accuracy of the model by reducing the impact of rare events. However, the choice of the smoothing parameter alpha is important and can impact the performance of the model. A small alpha value may result in overfitting, while a large alpha value may underfit the model.

```python
class NaiveBayesClassifier:
  priorProb = []
  classes = []
  class_counts = []
  attribute_values = []
  attribute_class_probs = []
  alpha = 1
    
  def __init__(self):
    pass

  def calcClass(self, daf): 
    self.classes, self.class_counts = np.unique(daf['income'], return_counts=True)

  #calculate prior probabilities
  def calc_prior_probabilities(self, y):
    unique, counts = np.unique(y, return_counts=True)
    counts = counts/len(y)
    #print(type(counts[0]))
    data = np.array([unique, counts])
    dataset = pd.DataFrame({'Class': data[0, :], 'PriorProbability': data[1, :]})
    #print(type(dataset.iat[0,1]))
    dataset.PriorProbability = dataset.PriorProbability.astype(float)
    # print(dataset)
    self.priorProb = dataset
#     print(self.priorProb)

  def calc_conditional_prob(self, daf, alpha):
    for i in range(0, 14):
        unique, counts = np.unique(daf.iloc[:, i], return_counts=True) 
        self.attribute_values.append(unique)

        att_p = []
        for j in range(0, len(self.attribute_values[i])):
            probs = []
            for k in range(0, len(self.classes)):
                df1 = daf[daf.iloc[:, i] == self.attribute_values[i][j]]
                df2 = df1[df1['income'] == self.classes[k]]
                nk = df2.shape[0]
                att_count = counts[j] if self.attribute_values[i][j] in daf.iloc[:, i].values else 0 # check if attribute value is present in the data
                p = (nk + alpha) / (self.class_counts[k] + alpha * len(self.attribute_values[i]) + att_count)
                probs.append(p)
            att_p.append(probs)
        self.attribute_class_probs.append(att_p)


  def predict(self, ser):
    prob = []
    p = 1
    for i in range(0, len(self.classes)):
      p = self.priorProb.at[i, 'PriorProbability']
      for a in range(0, 14):
        for b in range(0, len(self.attribute_values[a])):
          if ser[a]==self.attribute_values[a][b]:
            #print(type(p), self.attribute_class_probs[a][b][i])
            p = p * self.attribute_class_probs[a][b][i]
      prob.append(p)
    
    maxi = 0
    maxval = 0
    for i in range(0, len(self.classes)):
      if prob[i]>maxval:
        maxi = i
        maxval = prob[i]
#     print(prob, maxval, maxi, self.classes[maxi])
    return self.classes[maxi]


  def calc_df_classes(self, daf):
    pred = []
    for i in range(0, len(daf)):
      pred.append(self.predict(daf.iloc[i]))
    return pred

  #accuracy
  def calc_accuracy(self, pred, actual):
    count = 0
    for i in range(0, len(pred)):
      if pred[i]==actual[i]:
        count = count+1
    return count/len(pred)
```
